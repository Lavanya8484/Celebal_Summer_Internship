```python
from sklearn.datasets import load_iris
import pandas as pd

#iris dataset
iris_raw = load_iris()
iris = pd.DataFrame(data=iris_raw.data, columns=iris_raw.feature_names)
iris["species"] = pd.Categorical.from_codes(iris_raw.target, iris_raw.target_names)

# Re-naming columns
iris.columns = ['sepal_length', 'sepal_width', 'petal_length', 'petal_width', 'species']

iris.head()
```

```python
import matplotlib.pyplot as plt

species_counts = iris['species'].value_counts()

# Pie chart
plt.figure(figsize=(6, 6))
plt.pie(species_counts, labels=species_counts.index, autopct='%1.1f%%', startangle=140)
plt.title("Pie Chart - Species Distribution")
plt.show()
```

```python
import seaborn as sns

# Histogram
plt.figure(figsize=(8, 5))
sns.histplot(iris['petal_length'], bins=20, kde=True, color="red")
plt.title("Histogram - Petal Length")
plt.xlabel("Petal Length")
plt.ylabel("Frequency")
plt.show()
```

```python
# Line graph
plt.figure(figsize=(10, 5))
sns.lineplot(data=iris, x=iris.index, y="sepal_length", hue="species")
plt.title("Line Graph - Sepal Length by Index")
plt.xlabel("Index")
plt.ylabel("Sepal Length")
plt.grid(True)
plt.show()
```

```python
# Box plot
plt.figure(figsize=(8, 5))
sns.boxplot(data=iris, x="species", y="sepal_width", palette="Set2")
plt.title("Box Plot - Sepal Width by Species")
plt.show()
```
